```python
# Задача 1.
# Ответь на вопрос: игры каких жанров были наиболее популярны до 2000 года, а какие после?
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')
df = pd.read_csv("/content/vgsale_1.csv")

df['Year'] = pd.to_datetime(df['Year'], errors='coerce').apply(lambda x: str(x).split('-')[0] if x != np.nan else np.nan) # очищаю столбец даты

df_after = df.loc[df['Year'] > '2000']
df_before = df.loc[df['Year'] <= '2000']

after_2000 = (df_after['Genre']).value_counts().head(3)
before_2000 = (df_before['Genre']).value_counts().head(3)

print("После 2000-х популярны:")
print(after_2000, '\n')
print("До 2000-х популярны:")
print(before_2000)
```

```python
# Задача 1.
# Оцени популярность жанров по количеству выпущенных игр и по объему продаж по всему миру.
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns

warnings.filterwarnings('ignore')
df = pd.read_csv("/content/vgsale_1.csv")

lang_df = pd.DataFrame(df['Genre'].value_counts())

lang_df['Global_Sales'] = lang_df.index
lang_df.columns = ['Global_Sales', 'Genre']

plt.figure(figsize=(12,5))
plt.title("Популярность жанров по объему продаж")
sns.barplot(x='Genre', y='Global_Sales', data=lang_df.iloc[0:-1])
plt.show()
```

```python
# Задача 1.
# Отобрази на графике общее число видеоигр, выпущенных в каждом году.
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns

warnings.filterwarnings('ignore')
df = pd.read_csv("/content/vgsale_1.csv")


df['Year'] = (df['Year']).apply(lambda x: str(x).split('.')[0])
df['Year']

```

```python
lang_df = pd.DataFrame(df['Year'].value_counts())

lang_df['Global_Sales'] = lang_df.index
lang_df.columns = ['Global_Sales', 'Year']

plt.figure(figsize=(12,5))
plt.title("Выпуск игр по годам")
sns.barplot(x='Year', y='Global_Sales', data=lang_df.iloc[0:-1])
plt.tick_params(axis='x', labelrotation=45)
plt.show()
```

```python
# Задача 1.
# отобрази на круговых диаграммах доли суммарного объема продаж с 1980 г. до 2000 г.
# и с 2000 г. до 2020 г. в Северной Америке, Европе, Японии (также для построения корректных диаграмм используйте столбец “Other Sales”)
# от объема продаж по всему миру.

import pandas as pd
import numpy as np
import warnings
import plotly.graph_objects as go

warnings.filterwarnings('ignore')
df = pd.read_csv("/content/vgsale_1.csv")

df['Year'] = pd.to_datetime(df['Year'], errors='coerce').apply(lambda x: str(x).split('-')[0] if x != np.nan else np.nan) # очищаю столбец даты
```

```python
df.head(3)
```

```python
before_2000 = df[df['Year'] <= '2000']
sales_bef_2000 = before_2000.groupby('Genre')['Global_Sales'].sum()
sales_bef_2000
```

```python
import plotly.graph_objects as go
fig = go.Figure(data=[go.Pie(values=sales_bef_2000, labels=df['Genre'])])
fig.show()
```

```python
after_2000 = df[df['Year'] > '2000']
sales_af_2000 = after_2000.groupby('Genre')['Global_Sales'].sum()
sales_af_2000
```

```python
import plotly.graph_objects as go
fig = go.Figure(data=[go.Pie(values=sales_af_2000, labels=df['Genre'])])
fig.show()
```
